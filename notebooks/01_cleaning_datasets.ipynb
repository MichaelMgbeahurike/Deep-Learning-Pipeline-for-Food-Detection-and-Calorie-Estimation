```python
dish1 = pd.read_csv(path1, on_bad_lines='warn', engine='python')
dish2 = pd.read_csv(path2, on_bad_lines='warn', engine='python')
```


This dataset contains metadata for dishes, likely with detailed nutritional information.

The headers describe the data structure:

*   **Main Dish Information:**
    *   `Dish ID`
    *   `Total Calories`
    *   `Total Mass`
    *   `Total Fat`
    *   `Total Carbohydrates`
    *   `Total Protein`

*   **Ingredient Information (repeated for each ingredient in a dish):**
    *   `Ingredient ID`
    *   `Ingredient Name`
    *   `Grams`
    *   `Calories`
    *   `Fat`
    *   `Carbohydrates`
    *   `Protein`


```python
dish = pd.concat([dish1, dish2], ignore_index=True)
colCount = dish.shape[1]
maxIng = (colCount - 6) // 7
count = 6
headers = [
    "dish_id",
    "total_calories",
    "total_mass",
    "total_fat",
    "total_carbohydrates",
    "total_protein",
]
for i in range(1, maxIng+1):
  count += 7
  headers += [
      f"ingr_{i}_id",
      f"ingr_{i}_name",
      f"ingr_{i}_grams",
      f"ingr_{i}_calories",
      f"ingr_{i}_fat",
      f"ingr_{i}_carbohydrates",
      f"ingr_{i}_protein",
  ]

print(len(dish.columns))
print(dish.shape[1])
print(count)
print(maxIng*7 + 6)
dishes = pd.concat([dish1,dish2], ignore_index=True, headers = headers)
```

```python
keywords = ["oil","butter,ghee","salt","pepper","spice",
            "seasoning","vinegar","soy sauce","sauce",
            "dressing","marinade","syrup","honey","sugar",
            "broth","stock","stock","water","baking powder",
            "baking soda","yeast","extract","vanilla"
]
pattern = "|".join(re.escape(k) for k in keywords)

hits = dish[dish['ingr_name'].str.contains(pattern, case=False, na=False)].copy()

dish_hits = (
    hits.groupby('dish_name')["ingr_name"]
    .apply(lambda s: ", ".join(sorted(set(s))))
    .reset_index(name="hits")
)

dish_hits["hit_count"] = dish_hits["hits"].str.count(", ") + 1
dish_hits.head(20)


```
